```python
# Import numpy library
import numpy as np

# Function to calculate the transfer matrix for all layers
def TransferMatrix(MF, VM, DM, energy):
    # initialize transmission fraction vector
    tfrac = np.zeros( len(energy) )
    
    # Define constants
    # reduced Planck constant
    hbar = 1.0546E-34

    # electron mass
    mo = 9.1095E-31

    # electron charge
    q = 1.6022E-19

    # useful parameter
    s = (2 * q * mo * 1E-18) / (hbar**2)
    
    # Outer Loop: loop through all "m" energy values
    for m in range(len(energy)):
        # Step 1: Initialize transfer boundary condition matrix for first layer
        
        # Calculate "k" for layer "1"
        k1squared = s * MF[0] * (energy[m] - VM[0])
        if k1squared < 0:
            k1 = 1j * np.sqrt(np.absolute(k1squared))
        else:
            k1 = np.sqrt(k1squared)
        
        # Calculate "k" for layer "2"
        k2squared = s * MF[1] * (energy[m] - VM[1])
        if k2squared < 0:
            k2 = 1j * np.sqrt(np.absolute(k2squared))
        else:
            k2 = np.sqrt(k2squared)
        
        # Calculate "delta" for layer "1"
        if k1 == 0 or MF[1] == 0:
            delta1 = np.inf
        else:
            delta1 = (k2 / k1) * (MF[0] / MF[1])
        
        # Begin to construct transfer matrix by creating boundary condition matrix between layer 1 and 2
        trans = Calculate_Boundary_Condition_Matrix(delta1)
        
        # Inner Loop: loop through all remaining layers "n" of the structure (1 through final-1)
        for n in range(1, len(MF) - 1):
            # Create the propagation matrix and boundary condition matrix for each layer and multiply to update the transfer matrix
            # Calculate "k" for layer "n"
            
            knsquared = s * MF[n] * (energy[m] - VM[n])
            if knsquared < 0:
                kn = 1j * np.sqrt(np.absolute(knsquared))
            else:
                kn = np.sqrt(knsquared)
            
            # Calculate "k" for next layer
            kn1squared = s * MF[n + 1] * (energy[m] - VM[n + 1])
            if kn1squared < 0:
                kn1 = 1j * np.sqrt(np.absolute(kn1squared))
            else:
                kn1 = np.sqrt(kn1squared)
            
            # Calculate "delta" for layer "n"
            if kn == 0 or MF[n+1] == 0:
                deltan = np.inf
            else:
                deltan = (kn1 / kn) * (MF[n] / MF[n+1])
            
            # Multiply to the running matrix product "trans" the propagation and boundary condition matrix for layer "n"
            LayerMatrixn = Calculate_Propagation_Matrix(kn, DM[n]).dot(Calculate_Boundary_Condition_Matrix(deltan))
            
            trans = trans.dot(LayerMatrixn)

        # Calculate transmission fraction given the resulting transfer matrix for energy "m"
        tfrac[m] = 1 - np.square(np.absolute((trans[1, 0]))) / np.square(np.absolute((trans[0, 0])))
        print("m = ", m, "tfrac[m] = ", tfrac[m], "energy[m] = ", energy[m])
    
    return tfrac

# Function to compute the boundary condition matrix between layers "n" and "n+1", using "delta" for layer n
def Calculate_Boundary_Condition_Matrix(deltan):
    bcMatrix = (1 / 2) * np.array([[1 + deltan, 1 - deltan], [1 - deltan, 1 + deltan]])
    
    return bcMatrix


# Function to compute the propagation matrix for layer "n" using the potential, effective mass, and thickness of this layer
def Calculate_Propagation_Matrix(kn, dmn):
    propMatrix = np.array([[np.exp(-1j * kn * dmn), 0], [0, np.exp(1j * kn * dmn)]])
    
    return propMatrix
```

```python
# Define default values for transmission function
# Effective masses in each layer
MF = np.array([1, 1, 1, 1, 1])

# Potential in each layer
VM = np.array([0, 0.9, 0, 0.9, 0])

# Thickness of each layer (note: first and last values for infinite layers are not used)
DM = np.array([0, 0.3, 0.7, 0.3, 0])

# Vector of energies of interest
energy = np.linspace(0, 2, num = 50)

import time
start = time.time()

# Calculate transmission fractions
tfrac = TransferMatrix(MF, VM, DM, energy)

end = time.time()
print("Time to calculate transmission fractions = ", end - start)
```

```python
# Import matplot library
import matplotlib.pyplot as plt
%matplotlib inline

# Create transmission-vs-energy plot
plot = plt.plot(energy, tfrac)

# Define axis labels and plot title
plt.title('Transmission vs Energy Plot')
plt.xlabel('Energy')
plt.ylabel('Transmission Fraction')

# Set axis limits
plt.xlim(0, 2)
plt.ylim(0, 1)

plt.show()
```
